```python
import cv2
import numpy as np
from pathlib import Path

AER850_DIR = Path("/content/drive/MyDrive/AER850")

MOTHERBOARD_PATH = AER850_DIR / "motherboard_image.JPEG"

OUTPUT_DIR = AER850_DIR / "Methodology_Images"
OUTPUT_DIR.mkdir(parents=True, exist_ok=True)

print("Using input image:", MOTHERBOARD_PATH)
print("Saving outputs to:", OUTPUT_DIR)

# Original image
motherboard_bgr = cv2.imread(str(MOTHERBOARD_PATH))
if motherboard_bgr is None:
    raise FileNotFoundError(f"Could not read image at: {MOTHERBOARD_PATH}")

print("Loaded image shape:", motherboard_bgr.shape)

# Overall Process

# gray
motherboard_gray = cv2.cvtColor(motherboard_bgr, cv2.COLOR_BGR2GRAY)

# GaussianBlur

blurred_gray = cv2.GaussianBlur(motherboard_gray, (7, 7), 0)

# Canny edges
edge_map = cv2.Canny(blurred_gray, 30, 120)

kernel = np.ones((21, 21), np.uint8)
closed_edges = cv2.morphologyEx(edge_map, cv2.MORPH_CLOSE, kernel)

# Fingding countoursPCB
contours, _ = cv2.findContours(closed_edges, cv2.RETR_EXTERNAL, cv2.CHAIN_APPROX_SIMPLE)
if not contours:
    raise RuntimeError("No contours found in closed_edges image.")

pcb_contour = max(contours, key=cv2.contourArea)

# Build mask
pcb_mask = np.zeros(motherboard_gray.shape, dtype=np.uint8)
cv2.drawContours(pcb_mask, [pcb_contour], contourIdx=-1, color=255, thickness=-1)

# extracting the PCB
pcb_segment = cv2.bitwise_and(motherboard_bgr, motherboard_bgr, mask=pcb_mask)

# 4. save all the images to the drive

cv2.imwrite(str(OUTPUT_DIR / "pcb_original.png"), motherboard_bgr)
cv2.imwrite(str(OUTPUT_DIR / "pcb_gray.png"), motherboard_gray)
cv2.imwrite(str(OUTPUT_DIR / "pcb_edges.png"), edge_map)
cv2.imwrite(str(OUTPUT_DIR / "pcb_closed_edges.png"), closed_edges)
cv2.imwrite(str(OUTPUT_DIR / "pcb_mask.png"), pcb_mask)
cv2.imwrite(str(OUTPUT_DIR / "pcb_extracted.png"), pcb_segment)

print("\nDone. Saved:")
for name in [
    "pcb_original.png",
    "pcb_gray.png",
    "pcb_edges.png",
    "pcb_closed_edges.png",
    "pcb_mask.png",
    "pcb_extracted.png",
]:
    print("  -", OUTPUT_DIR / name)
```
